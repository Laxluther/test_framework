# SQLite DB Editor
Use this notebook for manual data exploration and updates of `results.db` if the UI override is not enough.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('results.db')
pd.set_option('display.max_columns', None)
```

## 1. View Batch Runs

```python
df_batches = pd.read_sql_query("SELECT * FROM batch_runs", conn)
df_batches.head()
```

## 2. View Individual Test Results

```python
df_tests = pd.read_sql_query("SELECT id, batch_id, round_no, conversation_no, grades_passed, assumptions_score FROM test_results", conn)
df_tests.head()
```
